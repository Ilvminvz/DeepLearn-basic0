```python
import tensorflow as tf
```

```python
from tensorflow import keras
```

```python
fashion = keras.datasets.fashion_mnist #fashion_mnist is a function within tensorflow
```

```python
(features_train, label_train), (features_test, labels_test) = fashion.load_data()
```

```python
features_train.shape
```

```python
features_test
```

```python
label_train
```

```python
label_names = ['T-Shirt','Trouser','Pullover','Dress','Coat','Sandal','Shirt','Sneaker','Bag','Ankle boot']
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.imshow(features_train[0])
plt.show()
```

```python
#normalise
features_train = features_train / 255.0
features_test = features_test / 255.0
```

```python
#features_train[0]
```

```python
for i in range(0,25):
  plt.figure()
  plt.imshow(features_train[i])
  plt.xlabel('features_name==label_name')
  plt.show()
```

```python
plt.figure(figsize=(10,10))
for i in range(25):
  plt.subplot(5,5,i+1)
  plt.xticks([])
  plt.yticks([])
  plt.imshow(features_train[i])
  plt.xlabel(label_names[label_train[i]])
plt.show()
```

```python
model = keras.models.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy']
              )
```

```python
model.fit(features_train, label_train, epochs=10)
```

```python
loss, accuracy = model.evaluate(features_train, label_train)
```
